# 1. Motivation

#### Our project investigates how air pollution patterns differ in major cities around the world. Initially, the project started with a narrower focus on Dubai, where we wanted to compare pollution patters across a few monitoring locations inside Dubai. However, after exploring the data and receiving feedback, we realized that focussing purely on the Dubai story was too limited for the type of narrative we wanted to build. The different sensor locations within Dubai were not strong enough to support a full data story on their own.

#### After considering our options on what to do, we expanded the project into a broader global comparison. Instead of asking whether Dubai has different pollution patterns internally, we now ask how different cities around the world compare in terms of air-quality behavior. Dubai is still included, but only as one city in a wider set of global cities.

#### The main goal of our project is not just to rank the cities from most to least polluted. If we take just a single average value, that would hide important differences in seasonoality, short pollution episodes, composition of the particles, weather context, and the burden of exceeding the daily limits. Our project is focused on the idea that every city has an air-quality “fingerprint.”

## **What is the dataset?**

#### The dataset is a combination of air-quality measurements from OpenAQ with historical weather data from Open-Meteo. We get the PM2.5 and PM10 measurements from monitoring locations in the 15 cities using OpenAQ. PM2.5 is comprised of smaller, fine particle matter, while PM10 is made of larger particles. To give our project more depth we added weather variables including temperature, humidity, precipitation, wind speed, wind direction, cloud cover, and surface pressure. These weather variables were obtained from Open-Meteo.

## **Why this dataset?**

#### We chose this dataset because air pollution is an important global issue and visually interesting at the same time. PM2.5 and PM10 vary across the cities, seasons, weather conditions, and pollution episodes. This gave us an opportunity for a data visualization project because the story cannot be understood from one number or variable alone. The combination of air-quality and weather data also allowed us to move further than simple city rankings and investigate broader patterns within.

## **What is the goal for the end user’s experience?**

#### The goal of the website is to help a non-technical reader understand that air pollution is not just about city averages. We wanted users to explore how cities differ over time, rankings change month by month, how often health-based guidlines were exceeded, and how the particle compositions vary. The website is designed as a guided visual story, while this notebook provides a more in depth explanation to the technical work behind it.

# 2. Basic Stats

#### The raw OpenAQ data was collected at a local monitoring station level. Since each city has multiple monitoring stations, there is quite a few measurements for the same city and hour. To make the data usable for comparison, we processed it into different levels: raw hourly pollutant records, location-hourly records, city-hourly records, city-daily records, and city-monthly records.

#### The raw dataset contained more than 700,000 hourly records across the selected cities and sensors. We chose to use 2025 as the common analysis year because of the recency and it is also the most commonly recorded year across the different city sensors. Sticking to one common year between the different cities made the comparison fairer and easier to explain.

### For city selection:

#### We initially considered 18 cities that had different urban and environmental contexts. After filtering through and checking data availability and quality, we selected 15 cities for the main analysis:

#### Bangkok, Beijing, Delhi, Dhaka, Dubai, Jakarta, Lahore, London, Los Angeles, Mexico City, New York, Riyadh, Seoul, Singapore, and Tokyo.

#### Abu Dhabi, Cape Town, and Kuwait City were removed from the main analysis because their coverage was not strong enough or did not align well with the 2025 comparison window.

## **Data Cleaning and Preprocessing**

#### Data cleaning was necessary because some of the pollutant values were clearly invalid. For example, we found extremely high values such as 9999 or values above 5000 which is not realistic. These PM2.5 measurements were likely missing-values or sensor errors. To account for a rare unfortunate problem with a sensor, we treated PM2.5 values below 0 or above 1000 as invalid, and PM10 values below 0 or above 2000 as invalid. These values were replaced with missing values before the daily and monthly datasets were rebuilt.

#### For many of the visual comparisons, we used the median for PM2.5 instead of mean. Median values are better to use when there are outliers from short spikes and sensor errors. When we did the WHO analysis however, we used daily mean PM2.5 because the official daily WHO guidline is based on a 24-hour mean.

# 3. Data Analysis

#### The main analytical idea of the project is that city averages hide important variation. A city with a moderate to low yearly average can still have many high-pollution days, while another city may have a high average but a more predictable seasonal pattern. To find out what the air pollution situation in these cities really was, we analyzed the data from several perspectives instead of relying on a single summary statistic.

### **Monthly PM2.5 comparison**

#### The first analysis compared monthly median PM2.5 across the selected cities. This gave us a broad overview of how pollution changed across 2025. We used monthly medians because they are easier to compare than hourly values since the monthly ones are less sensitive to short extreme spikes/outliers. The analysis showed that air pollution is not a fixed number throughout the year. Some cities had a consistent high PM2.5 level, while others experienced significant changes from month to month, showing that air quality is more dynamic than meets the eye.

### **City-month heatmap**

#### The city-month heatmap was used to reveal seasonality. Instead of showing only which cities are more polluted on average, the heatmap shows when pollution is higher or lower during the year. This helps identify whether a city’s air-quality problem is stable, seasonal, or episode-driven.

### **Monthly ranking transitions**

#### We then ranked cities by monthly median PM2.5. The purpose of this analysis was to show that rankings are not fixed. Some cities remain consistently high, while others move up or down depending on the month. This supports the main argument that a single annual ranking can hide important dynamics that change throughout the year.

### **WHO exceedance analysis**

#### To connect the analysis to health relevance, we calculated how often each city exceeded the WHO daily PM2.5 guideline. For this part, we used daily mean PM2.5, since the guideline is defined using a 24-hour mean. The exceedance dot matrix makes the health burden visible and shows just how many days this guidline is breached.

### **Particle-profile analysis**

#### PM2.5 and PM10 describe different particle-size ranges. To investigate this, we calculated the PM2.5 / PM10 ratio for city-days where both pollutants were available. A higher ratio suggests that fine particles dominate, while a lower ratio suggests a larger contribution from coarse particles. This helped us compare not only how polluted cities are, but also what type of particle profile they tend to have.

### **Weather-episode fingerprint**

#### We added on to the air-quality data by including weather variables from Open-Meteo. Weather was not used to prove causation, but to add context to why pollution episodes/extremeties take place. For example, wind speed and particle ratios can help distinguish between different types of unusually extreme pollution. The weather-episode fingerprint visualizes a combination of wind speed, PM2.5 / PM10 ratio, PM10 level, and episode type in one plot.

### **City fingerprint map**

#### Finally, we created a city fingerprint map using multiple features, including pollution burden, exceedance frequency, seasonality, particle profile, and weather-linked factors. The goal was to group cities based on their overall behaviour in air-quality rather than taking the simple one-dimensional approach of ranking them by average PM2.5.

# 4. Genre

#### The website follows a magazine-style interactive data story. It is not purely exploratory, because the user is guided through the argument that city averages hide important pollution patterns. It is not a fully linear slideshow either, since several of the visulizations allow the user to explore the different cities, months, and patterns by interacting with them.

#### In terms of Segel and Heer’s data-storytelling genres, our project is closest to an interactive slideshow with a magazine-style narrative. The website has multiple sections, explanatory text, captions, and plots with limited interactiveness which guide the reader through the story.

### **Visual Narrative Tools**

#### For visual narrative, we used a consistent visual style with section headers and captions. The website has a navigation menu on the left side of the page so the reader can move back and forth between different parts of the story. The visualizations are embedded directly into the page so the reader can interact with the data without having to disrupt the flow of the datastory.

### **Narrative Structure**

#### For narrative structure, we used a guided ordering of scenes. The story begins with the broad global comparison of cities. Then it moves on to seasonality, rankings, exceedance burden, particle profiles, weather context, and finally city fingerprints. This order helps the reader build their understanding gradually and move from simple comparisons to more complex interpretations.

# 6. Discussion

#### One strength of the project is that it evolved based on what the data could actually support. Instead of forcing a story based on purely polution in Dubai, we expanded the scope to a global comparison when the data suggested that this would create a stronger narrative. This made the final project more interesting and more appropriate for visual analysis.

#### Another strength is that the project combines multiple levels of analysis: monthly patterns, daily exceedance burden, particle profiles, weather context, and city-level clustering. This gives the reader a more complete understanding of air pollution than a simple ranking would.

#### A limitation we faced is that monitoring coverage differs between the cities. Even though we selected a 2025 as a common time window and filtered sensor availability, some cities still have better coverage than others. Another limitation is that city-level data can hide variation on the neighborhood/local area level. Since the final project compares cities globally, we chose city-level values for consistency, but this means the analysis does not capture the local level differences within each city.

#### The visualizations can reveal patterns, similarities, and unusual episodes, but they cannot fully explain the causes behind them. For example, the PM2.5/PM10 ratio helps us understand the differences in particle composition but it doesn't identify the exact sources the pollution comes from. To do that, we would need additional data such as emission numbers from traffic patterns, land use, industrial activity, or wildfire events.

#### A final point is that Dubai still appears in the project, but its role changed. At the beginning we intended to use Dubai as the main focus. After exploring the data further, we decided that a global comparison would create a stronger data story. Therefore, Dubai became an example city to demonstrate how local timing and weather add detail to pollution comparisons worldwide. This allowed us to keep our original work and idea for the project without making the whole story depend on Dubai alone.

#### Overall, the project does successfully show that air pollution is multi-dimensional. The same pollution value for a city can look different depending on whether we focus on monthly medians, daily exceedances, particle ratios, weather episodes, or fingerprint similarity.
